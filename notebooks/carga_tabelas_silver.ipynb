```python
# Importação das bibliotecas
from sqlalchemy import create_engine
from datetime import datetime, timedelta
import time
import pandas as pd
from db import *

engine = get_postgres_engine()

# Lê a tabela diretamente para um DataFrame
df = pd.read_sql('SELECT * FROM public."TB_VENDAS_EMPRESA_B_BRONZE"', engine)
# Exibe os primeiros registros
print(df.head())
```

```python
import urllib

engine2 = get_sqlserver_engine()

df2 = pd.read_sql('SELECT * FROM TB_VENDAS_EMPRESA_A_BRONZE', engine2)

# Converte para string e substitui ',' por '.'
df2['ValorReais'] = df2['ValorReais'].astype(str).str.replace(',', '.', regex=False)

df2.drop_duplicates(inplace=True)

print(df2.head())
```
