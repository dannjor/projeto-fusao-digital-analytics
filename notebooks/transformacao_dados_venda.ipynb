Enriquecendo a tabela de vendas com os dados da cotação

```python
# Importação das bibliotecas
from sqlalchemy import create_engine
from datetime import datetime, timedelta
import time
import urllib
import pandas as pd
from db import *

engine = get_postgres_engine()
engine2 = get_sqlserver_engine()

df = pd.read_sql("""SELECT E.NomeProduto, E.CategoriaProduto, CAST(E.DataVenda AS DATE) DataVenda, E.EstadoFilial, E.ValorTotalVenda, 
ISNULL(A.CotacaoVenda, (SELECT MAX(CotacaoVenda) FROM TB_VENDAS_AUXILAR)) AS CotacaoDolar
FROM TB_VENDAS_EMPRESA_A_EMPRESA_B E
LEFT JOIN TB_VENDAS_AUXILAR A ON A.DataCotacao = E.DataVenda""", engine2)

# Exibe os primeiros registros
print(df.head())
```

Gravando os dados do dataframe nas tabelas finais

```python
# Envia os dados para o SQL Server
df.to_sql("TB_VENDAS_FINAL", engine, if_exists="replace", index=False)
df.to_sql("TB_VENDAS_FINAL", engine2, if_exists="replace", index=False)

print("✅ Dados carregados com sucesso no SQL Server e Postgress!")
```
